```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

df = pd.read_csv('twitter_dataset.csv')
df['Timestamp'] = pd.to_datetime(df['Timestamp']) #converting string to datetime object
df.head(3)
```

```python
analyzer = SentimentIntensityAnalyzer() #VADER ONNN!!!!!


scores = df['Text'].apply(analyzer.polarity_scores) # assigns different scores to the text in the data, basically does that to each row of the text column  
sentiment_df = pd.DataFrame(scores.tolist()) # convert list of dicts to df 
df[['pos', 'neg', 'neu', 'compound_score']] = sentiment_df[['pos', 'neg', 'neu', 'compound']].values
df[['Text', 'pos', 'neg', 'neu', 'compound_score']].head()
```

```python
def get_label(compound): # function to give lavel to the score, less than -0.05 is negative, more than 0.05 is positive, in between is neutral
    if compound >= 0.05:
        return 'positive'
    elif compound <= -0.05:
        return 'negative'
    else:
        return 'neutral'

df['sentiment_label'] = df['compound_score'].apply(get_label) #now gets the label for each row's compound score
print(df['sentiment_label'].value_counts())
```

```python
df.to_csv('tweets_with_sentiment.csv', index=False) #makes new csv with new columns for the sentiment scores and labels
print('Done! Columns:', list(df.columns))
```
